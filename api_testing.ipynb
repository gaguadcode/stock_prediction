```python
import requests
import os
from dotenv import load_dotenv 
load_dotenv()
api_key = os.getenv("api_key")
# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=BRENT&interval=monthly&apikey=api_key'
r = requests.get(url)
data = r.json()
data = [{'month': entry['date'][:], 'price': entry['value']} for entry in data['data']]

print(data)
```

```python
import pandas as pd
#brent_price = [{'month': entry['date'][:], 'price': entry['value']} for entry in data['data']]
# Create a DataFrame
df = pd.DataFrame(data)
# Save DataFrame to CSV
df.to_csv('crude_oil_prices_brent.csv', index=False)
print(data)
data
```
